# Answer the key questions given below:

```python
# Calculate the average house price using .mean()
avg_house_price = hp['price'].mean()
print('The average house price is ' + str(avg_house_price))
```

```python
# Find the most common number of bedrooms and bathrooms
most_common_bedroom_number = hp['bedrooms'].mode()[0]
most_common_bathroom_number = hp['bathrooms'].mode()[0]
print('The most common bedroom number is ' + str((most_common_bedroom_number)))
print('The most common bathroom number is ' + str(most_common_bathroom_number))
```

```python
# Identify any correlation between the house price and other numeric features (e.g., sqft_living, waterfront, and view) using .corr() functions
numeric_features = ['sqft_living','waterfront', 'view']
correlation = hp[numeric_features + ['price']].corr(numeric_only = True)['price']
print(correlation)
```

```python
# Calculate the average price of waterfront and non-waterfront houses
waterfront_homes = hp[hp['waterfront'] == 1]
non_waterfront_homes = hp[hp['waterfront'] == 0]

avg_price_waterfront_homes = waterfront_homes['price'].mean()
avg_price_non_waterfront_homes = non_waterfront_homes['price'].mean()

print('The average price for waterfront homes is ' + str(avg_price_waterfront_homes))
print('The average price for non waterfront homes is ' + str(avg_price_non_waterfront_homes))
```

```python
# Determine the average price of houses with different numbers of floors
avg_price_by_floors = hp.groupby('floors')['price'].mean()
print(avg_price_by_floors)
```

```python
# Analyze the average price based on the house condition
avg_price_by_condition = hp.groupby('condition')['price'].mean()
print(avg_price_by_condition)
```
