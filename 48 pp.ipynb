```python
# In a new dataframe, provide the code that replaces the blank values with the average petal width for Setosa varieties. 
# (If you are having trouble with this, it may be best to break your work into multiple lines.)
ir2 = pd.read_csv('iris_dirty.csv')

avg_petal_width = ir2.loc[ir2['variety'] == 'Setosa', 'petal.width'].mean()

ir2.loc[(ir2['variety'] == 'Setosa') & (ir2['petal.width'].isna()), 'petal.width'] = avg_petal_width
ir2
```
