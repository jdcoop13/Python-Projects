```python
# Handle missing data using df.isnull().sum() to count null values in the DataFrame.  Fill missing values in the DataFrame using df.fillna().
print(tt.isnull().sum())
tt = tt.fillna('Unknown')
```

```python
# Conduct aggregate analysis using group-by operations: 2-3 aggregations
group1 = tt.groupby('Sex')['PassengerId'].count()
print(group1)

group2 = tt.groupby('Survived')['Pclass'].mean()
print(group2)
```

```python
# Visualize the data using basic plotting techniques: 2 charts
tt
```

```python
import matplotlib.pyplot as plt
```

```python
count_of_passengerid = tt['Sex'].value_counts()
count_of_passengerid.plot(kind = 'bar', x = 'Sex', y = count_of_passengerid, color = ('blue', 'red'))
plt.ylabel('Count of Passenger ID')
plt.show()
```

```python
avg_pclass = tt.groupby('Survived')['Pclass'].mean()
avg_pclass.plot(kind = 'bar', x = 'Sex', y = avg_pclass, color = ('orange', 'green'))
plt.ylabel('Avg Pclass')
plt.show()
```
